```python
input_video_path = 'd.mp4'
output_video_path = 'output_video.avi'

# Open the input video
cap = cv2.VideoCapture(input_video_path)

# Get video properties
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Define a video writer for output
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))
```

```python
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO inference with a confidence threshold of 51%
    results = model(frame, conf=0.50)  # Filter detections with confidence ≥ 0.51

    # Draw the detections on the frame
    annotated_frame = results[0].plot()  # Annotated frame with bounding boxes

    # Write the annotated frame to the output video
    out.write(annotated_frame)

    # (Optional) Display the frame in real-time
    cv2.imshow('YOLOv8 Inference', annotated_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):  # Press 'q' to quit
        break

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
import cv2
import requests
import random
from datetime import datetime
# Set up video capture
cap = cv2.VideoCapture("ras.mp4")  # Replace with your video file or camera index
out = cv2.VideoWriter(
    "output_video.avi",
    cv2.VideoWriter_fourcc(*"XVID"),
    30,  # Adjust FPS as needed
    (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))),
)

# YOLOv8 model (adjust as per your setup)
from ultralytics import YOLO

model = YOLO("best.pt")  # Load YOLO model (adjust as needed)

# API endpoint to add detections
API_URL = "http://192.168.1.105:5000/add_detection"
info = [
        {
            "city" : "Jendouba , TN",
            "address" : "Foret Tbarka"
        },
        {
            "city" : "Beja , TN",
            "address" : "Foret Beja"
        },
        {
            "city" : "Siliana  ,TN",
            "address" : "Djebal Chahid"
        }
    ]
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO inference with a confidence threshold of 50%
    results = model(frame, conf=0.50)

    # Draw the detections on the frame
    annotated_frame = results[0].plot()  # Annotated frame with bounding boxes
    now = datetime.utcnow()
    

   

# Format the date and time in the desired format
    formatted_time = now.strftime('%Y-%m-%dT%H:%M:%SZ')
    # Iterate through detections
    for result in results[0].boxes.data:
        # Each result contains [x1, y1, x2, y2, confidence, class_id]
        x1, y1, x2, y2, confidence, class_id = result.cpu().numpy()
        if confidence >= 0.50:  # Check confidence threshold
            # Prepare detection data
            data = random.choice(info)
            detection_data = {
                "coordinates" : [37.5875441,24.4475875],
                "city" : data["city"],
                "address" : data["address"],
                "confidence": float(confidence),
                "timestamp": str(formatted_time),
                "class_id": int(class_id),
            }

            # Send POST request
            try:
                response = requests.post(API_URL, json=detection_data)
                if response.status_code == 200:
                    print("Detection added successfully:", detection_data)
                else:
                    print("Failed to add detection:", response.status_code, response.text)
            except requests.exceptions.RequestException as e:
                print("Error sending POST request:", e)

    # Write the annotated frame to the output video
    out.write(annotated_frame)

    # (Optional) Display the frame in real-time
    cv2.imshow('YOLOv8 Inference', annotated_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):  # Press 'q' to quit
        break

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()
```
